#     Scale of Measurement :

1. Nominal Scale data
     - includes qulitative/categorical
     - eg : gender colors
     - order doesn't matters

2. ordinal scale data
     - ranking is imp
     - order matters
     - difference can't be measured
     - eg: [1->good  2->beter 3->bad]

3. interval scale data
     - order matters
     - difference can be measured 
     - ratio can't be measured
     - no '0' starting point
     - eg temperature {23,46,67} 
          >   46-23
          >   but we cant say on day two hotness was double than that on day one
          >   it can be negative too

4. ratio scale 
     - order matters 
     - difference can be measured (including ratio)
     - contains a '0' starting point
     - eg : marks scored by std [ 30, 90, 60 , 75, 40, 0 ]
          >   can measure 90 -30 ,... we say kisi kitna jyada mila
          >   90/30  we say mujhe tujhse 3 guna jyada mila / dugna mila hai
          >   a std can score min. 0

# Covariance and correlation

### Covariance


- cov(x,x) = var(x)

- x directly propotional to y ..... +ve covariance
- x inversly propotional to y ..... -ve covariance

Advantages :
- helps to find out relationship b/w x and y

Disadvantage :
- covariance doesn't have a specific limit value


### Pearson correlation coefficient 
- always lies b/w [-1 , 1]
- more towards the -1 more -ve corellation  and vice versa




### Spearman rank correlation 
- lies b/w [-1,1]
-  used in feature selection of ML
- here we do cal culation based on rankings not on actual valuse
- x -----> 7  8  6  5  3  1   
  R(x) -> 2  1  3  4  5  6


```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
df.cov()
```

```python
#  pearson correlation 
df.corr(method='pearson')
```

```python
#  spearman correlation 
df.corr(method='spearman')
```
