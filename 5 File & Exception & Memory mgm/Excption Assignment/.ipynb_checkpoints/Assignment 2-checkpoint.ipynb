```python
#Q2. Write a python program to print Python Exception Hierarchy.

def treeClass(cls, ind = 0):
    s = '-' * ind + cls.__name__
    print (s)        # print name of the class
    logging.info(s)
    for i in cls.__subclasses__():          # iterating through subclasses
        treeClass(i, ind + 5)


logging.info("Hierarchy for Built-in exceptions is : ")
print("Hierarchy for Built-in exceptions is : ")
treeClass(Exception)
```

```python
'''
1.ZeroDivisionError:
       This exception is raised when attempting to divide a number by zero.
'''

try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("Error: Division by zero")
    print("Error: Division by zero")
```

```python
"""
2.OverflowError:
       This exception is raised when the result of an arithmetic operation is too large 
       to be expressed within the range of the available numeric type.
"""
import sys

try:
    result = sys.maxsize + 1
except OverflowError:
    logging.error("Error: Result exceeds the maximum value")
    print("Error: Result exceeds the maximum value")
```

```python
"""
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class for exceptions that occur when a lookup or
indexing operation fails.
"""
#1.KeyError
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
except KeyError:
    logging.error("Error: Key not found in dictionary")
    print("Error: Key not found in dictionary")
#2.IndexError
my_list = [1, 2, 3]

try:
    value = my_list[3]
except IndexError:
    logging.error("Error: Index out of range")
    print("Error: Index out of range")
```

Q5. Explain ImportError. What is ModuleNotFoundError?


=>ImportError is an exception that is raised when an import statement fails to import a module.

=>Common reasons for ImportError include:
*The module or package you are trying to import does not exist or is misspelled.
*The module is not present in the current working directory or any of the directories in sys.path.
*The module depends on other modules that are not installed or not accessible.
*There might be syntax errors or other exceptions raised while importing the module.


=>ModuleNotFoundError is a specific subclass of ImportError that is raised when a module cannot be   found during the import process.
=>Example:-

```python
try:
    import python
except ModuleNotFoundError:
    logging.error("Error: Module not found")
    print("Error: Module not found")
```
