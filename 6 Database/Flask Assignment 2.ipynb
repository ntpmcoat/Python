```python
Q1. Explain GET and POST methods.
```

    GET Method
=> Used for retrieving data from a server.
=> Appends data to the URL in the form of query parameters.
=> Limited data capacity (around 2048 characters) due to URL length restrictions.
=> Data is visible in the URL, which may pose security risks.
=> Caching is possible, which can improve performance for repeated requests.
=> Should only be used for safe and idempotent operations (does not change server state).


    POST Method
=> Used for submitting data to a server to create or update a resource.
=> Sends data in the request body, not visible in the URL.
=> Can handle large amounts of data.
=> More secure than GET as data is not visible in the URL.
=> Not cached by default, reducing the likelihood of unintended repeats.
=> Can be used for non-idempotent operations (changes server state with each request).

=>Flask uses the "request" object to handle incoming HTTP requests in a web application.

=> HTTP requests are the means by which clients (e.g., browsers) communicate with the server to request resources or perform actions on the server.

=> Flask is a web framework for Python that simplifies building web applications. To handle incoming requests, Flask provides the "request" object.

=>The "request" object allows Flask to access data sent by clients, such as form data, query parameters, cookies, and HTTP headers.

=>It enables the web application to respond dynamically based on the data in the request, generating customized responses for each client.

=>The "request" object also enables handling various HTTP methods (e.g., GET, POST, PUT, DELETE) that clients use to interact with the server.

=>By using the "request" object, Flask allows developers to access and process client data efficiently and build interactive web applications.

```python
Q3. Why is redirect() used in Flask?
```

=> The `redirect()` function in Flask is used to redirect the client's browser to a different URL.

=>It helps in handling specific scenarios where you want to send the user to another page or route within the application after performing certain actions.
-
=>Common use cases include redirecting users after a successful form submission or to enforce a specific URL pattern for a route.

=>Instead of rendering a new page directly, `redirect()` triggers the client to make a new request to the specified URL.

=>This ensures proper handling of URLs and prevents issues related to duplicate form submissions or incorrect URL patterns.

=>Overall, `redirect()` is a convenient way to guide users to different parts of the application and improve user experience by managing navigation flows effectively.
