Q3. What is App routing in Flask? Why do we use app routes?

=>In Flask, app routing refers to the process of associating URL patterns with specific view functions, allowing the web application to respond to different URLs with appropriate content. 

=>These view functions handle incoming HTTP requests and generate the corresponding HTTP responses.

=>It is accomplished using the @app.route() decorator, which associates a URL with a Python function. 




***Here's why app routes are essential in Flask:

URL mapping: 
App routes allow us to define the URL structure of our web application and map each URL to a specific view or functionality.

Handling HTTP methods:
App routes can be configured to handle different HTTP methods (GET, POST, etc.), enabling different actions based on the type of request.

Separation of concerns: 
Routing helps to separate the concerns of different parts of the application, making it easier to manage and understand the codebase.

Dynamic content:
App routes enable the passing of variables and parameters within the URLs, allowing for dynamic content generation and personalized responses.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

=>In Flask, the url_for() function is used for URL building.

=>It generates a URL for the given endpoint and arguments by taking care of the dynamic parts and the special characters that are present in the URL.

=>This function is particularly useful when you want to avoid hardcoding URLs in your templates and code, making your application more maintainable and flexible

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, this is the homepage!'

@app.route('/user/<username>')
def show_user_profile(username):
    return f'Hello, {username}! This is your profile.'

if __name__ == '__main__':
    with app.test_request_context():
        # Generating URLs using url_for()
        print(url_for('index'))                        # Output: /
        print(url_for('show_user_profile', username='John'))  # Output: /user/John
    app.run(host="0.0.0.0")
```
