Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, c= '#E606FF', marker= 'x')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("random data")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])




=>The subplot() function is used to create multiple plots (subplots) within a single figure, allowing us to display different plots side by side or in a grid layout. This is useful when we want to compare different visualizations or display related data together.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(12, 2.5))

plt.subplot(1,4, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(1,4, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(1,4,3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(1,4 ,4)
plt.plot(x, y4)
plt.title("Line 4")


plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])




=> A bar plot is a graphical representation of data using rectangular bars.
=> Each bar's length or height is proportional to the value it represents.
=> Bar plots are used to compare categorical data or to display the frequency or count of different categories.

    Usage:
=> Comparing values of different categories.
=> Displaying data distribution.
=> Visualizing relationships between categorical variables

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("bar")

plt.subplot(1,2,2)
plt.barh(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("horizontal-bar")
plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


=> A box plot is a graphical representation of the distribution of a dataset.
=> It displays the median, quartiles, and potential outliers in the data
=> providing a visual summary of its central tendency and spread

    Box Plot is Used for:
=> Identify Outliers: Box plots help identify potential outliers in a dataset.
=> Compare Distributions: Box plots are useful for comparing the distribution of data across different groups or categories.
=> Show Central Tendency: Box plots show the median, providing insight into the central tendency of the data.
=> Visualize Distribution: They offer a visual representation of the data's distribution, highlighting key summary statistics.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Boxes')
plt.ylabel('Values')

plt.show()
```
