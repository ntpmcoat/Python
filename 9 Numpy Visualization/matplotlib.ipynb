```python
#  color can also be given directly by color code
colours = np.random.rand(50)  #generates random values for color 

size = 1000*np.random.rand(50)  
# generatrss random valu which will later be used for defining size of points


#  alpha--> controls intensity how dark or light color should be, its range is[0,1]

# marker -->which mark shoulde be used int plotting (x, v)

x= np.random.rand(50)
y =np.random.rand(50)
plt.scatter(x,y, c= '#E606FF', s = size, alpha=.8 , marker= 'v')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("random data")
```

```python
x= np.linspace(0,10 ,100)
y = np.sin(x)


plt.figure(figsize=(6,2))       # to control size 

plt.grid()                    #gives grid box in background

# --  denote dashed line
#  also can give color r/g/b
plt.plot(x,y,'--g')
```
